```python
import numpy as np

from scipy.stats import invgamma

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
x = np.linspace(0.0, 10, 1000)
a = 50
pdf = lambda x: invgamma.pdf(x, a=a, scale=(a + 1))
ax.plot(x, pdf(x), "r-", lw=2, label="invgamma pdf")
# vline at 1
ax.axvline(x=1, color="k", linestyle="--")
```

```python
# sample from invgamma
t = lambda: invgamma.rvs(a=a, scale=(a + 1))
t()
```
